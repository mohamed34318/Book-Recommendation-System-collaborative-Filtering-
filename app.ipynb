```python
rating_with_books = books_df_copy.merge(Ratings_df_copy,how="inner", on="ISBN")
rating_with_books.head()
```

```python
# number of rating for each book and user
ratings_books_numbers = rating_with_books.groupby(['Book-Title'])['Book-Rating'].count().reset_index()
ratings_books_numbers.columns = ['Book-Title' , 'Number of Ratings']
ratings_books_numbers = ratings_books_numbers.sort_values(by='Number of Ratings' , ascending=False)
ratings_books_numbers.head()
```

```python
# average rating per book
ratings_books_avg = rating_with_books.groupby(['Book-Title'])['Book-Rating'].mean(numeric_only=True).reset_index()
ratings_books_avg.columns = ['Book-Title' , 'avg of Ratings']
ratings_books_avg = ratings_books_avg.sort_values(by='avg of Ratings' , ascending=False)
ratings_books_avg.head()
```

```python
Popularity_df = ratings_books_numbers.merge(ratings_books_avg , on='Book-Title')
Popularity_df.head()
```

```python
Popularity_df = Popularity_df[Popularity_df['Number of Ratings']>=250].sort_values(by='avg of Ratings' , ascending=False)
Popularity_df.head()
```

```python
# Now Merge with user ID
popular_df=Popularity_df.merge(books_df_copy,on='Book-Title').drop_duplicates('Book-Title')
popular_df.head()
```

```python
educated_users = rating_with_books.groupby('User-ID')['Book-Rating'].count() > 200
educated_users = educated_users[educated_users].index
educated_users
```

```python
filter_rating = rating_with_books[rating_with_books['User-ID'].isin(educated_users)]
filter_rating.head()
```

```python
# get famous book for each one
fam_book = filter_rating.groupby('Book-Title')['Book-Rating'].count() > 50
fam_book = fam_book[fam_book].index
fam_book
```

```python
final_rating = filter_rating[filter_rating['Book-Title'].isin(fam_book)]
final_rating.head()
```

```python
final_rating.to_csv('books_ss.csv')
```

```python
pt = final_rating.pivot_table(index='Book-Title' , values= 'Book-Rating' , columns ='User-ID').fillna(0)
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity_score = cosine_similarity(pt)
```

```python
def recommend(book_name):
    index=np.where(pt.index==book_name)[0][0]
    similar_items=sorted(list(enumerate(similarity_score[index])),key=lambda x:x[1],reverse=True)[1:6]
    for i in similar_items:
        book_title = pt.index[i[0]]
        print(book_title)
```

```python
recommend('4 Blondes')
```

```python
import pickle

# Assuming `similarity_score` and `pt` (pivot table) are your model components
model_data = {
    'similarity_score': similarity_score,
    'pt': pt
}

# Save to a .pkl file
with open('book_recommendation_model.pkl', 'wb') as file:
    pickle.dump(model_data, file)
```
